```python
# clean data
#df.isnull().sum()
data = df[["Date","Category","Note","Amount","Income/Expense"]]
data.head()
```

```python
# add new expenses to dataset
def add_expense(date,category,note,amount,exp_type="Expense"):
    global data
    new_expense = pd.DataFrame({
        "Date":[date],
        "Category":[category],
        "Note":[note],
        "Amount":[amount],
        "Income/Expense":[exp_type]
    })
    print(new_expense)
    data = pd.concat([data,new_expense],ignore_index=True)
    data.to_csv("expense_data_1.csv",index=False)
    print("Expense added successfully!")


add_expense("2025-08-22 19:30", "Food", "Shawarma", 2500, "Expense")
add_expense("2025-08-23 08:00", "Subscriptions",
            "Netflix Monthly Plan", 4500, "Expense")
add_expense("2025-08-24 14:00", "Entertainment",
            "Outdoor Games with friends", 7000, "Expense")
```

```python
# view recently added expenses
def view_expenses(n=5):
    global data
    print(data.tail(n))

view_expenses(3)
```

```python
# summarize expenses by category
def summarize_expenses(by="Category"):
    global data
    summary = data.groupby(by)["Amount"].sum().reset_index()
    print(summary)
summarize_expenses()
```
